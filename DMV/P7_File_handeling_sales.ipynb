# Task -3 : Perform data cleaning operations, such as handling missing values, removing duplicates, or correcting inconsistencies. 

```python
# When you want to clean your data by removing rows or columns that have missing or invalid values.
csv.dropna()

# df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

# Parameters:
# axis:
# 0 (default) means drop rows with missing values.
# 1 means drop columns with missing values.


# how:
# 'any' (default) means drop rows/columns that contain any NaN values.
# 'all' means drop rows/columns that contain only NaN values.


# thresh:
# Specifies a minimum number of non-NaN values required to keep the row/column.
#  For example, thresh=2 will keep rows/columns that have at least 2 non-NaN values.


# subset:
# Specify a subset of columns/rows to consider when dropping NaN values.
#  For example, subset=['col1', 'col2'] will only consider these two columns when deciding which rows to drop.
```

```python
csv.drop_duplicates()
# df.drop_duplicates() function in pandas is used to remove duplicate rows from a DataFrame.

# df.drop_duplicates(subset=None, keep='first', inplace=False)

# Parameters:
# subset:
# Specifies which columns to consider for identifying duplicates.
# If None (default), it considers all columns to identify duplicates.
# If you want to check duplicates based on certain columns, you can pass a list of column names (e.g., ['col1', 'col2']).


# keep:
# Determines which duplicates to keep:
# 'first' (default): Keeps the first occurrence and removes subsequent duplicates.
# 'last': Keeps the last occurrence and removes the earlier duplicates.
# False: Removes all duplicates, leaving only unique rows.


# inplace:
# If True, modifies the original DataFrame in place, removing duplicates without returning a new DataFrame.
# If False (default), returns a new DataFrame with duplicates removed.
```

# Task -4 : Convert the data into a unified format, such as a common dataframe or data structure, to enable seamless analysis. 

```python
# Assuming 'id' is a common identifier across datasets
unified_data = pd.concat([csv, ed, json_data], ignore_index=True)
unified_data

# pd.concat() function in pandas is used to concatenate or combine two or more DataFrames or Series along a particular axis (rows or columns). 
# It is a powerful function to merge data from multiple sources into a single DataFrame.

# pd.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)


# Parameters:
# objs:
# A sequence (list or tuple) of DataFrames or Series to be concatenated.
# Example: [df1, df2] or (df1, df2)


# axis:
# 0 (default) means concatenate along the rows (vertically), i.e., stack DataFrames on top of each other.
# 1 means concatenate along the columns (horizontally), i.e., add DataFrames side by side.


# join:
# Defines how to handle indices (rows or columns) that do not match across the objects being concatenated:
# 'outer' (default): Takes the union of the indices (keeps all rows/columns from all DataFrames).
# 'inner': Takes the intersection of the indices (keeps only common rows/columns).


# ignore_index:
# If True, it resets the index of the resulting DataFrame. It will create a new integer-based index.
# If False (default), the original indices are kept.


# keys:
# A sequence of keys to use as the new outermost level of the index. Useful for creating a hierarchical index.


# verify_integrity:
# If True, checks for duplicate indexes in the concatenated DataFrames. If any duplicates are found, it raises an error.


# sort:
# If True, it sorts the columns when concatenating along axis=1. Default is False.


# copy:
# If True (default), it copies the data. If False, it tries to avoid copying if possible.
```

# Task -5 : Analyze the sales data by performing descriptive statistics, aggregating data by specific variables, or calculating metrics such as total sales, average order value, or product category distribution. 

```python
total_sales = unified_data['Sales'].sum()
print(f"Total Sales: {total_sales}")
category_sales = unified_data.groupby('Invoice ID')['Sales'].mean()
```

# Task -6 : Create visualizations, such as bar plots, pie charts, or box plots, to represent the sales data and gain insights into sales trends, customer behavior, or product performance. 

```python
category_counts = unified_data['Sales'].value_counts()
# category_counts.plot(kind='bar',)
category_counts.plot(kind = "box",subplots = True,layout = (3,2),figsize=(15,20))

plt.title('Product Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
