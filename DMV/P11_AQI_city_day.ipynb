```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot AQI trend over time
def plot_aqi_trend(df):
    plt.figure(figsize=(10, 6))
    plt.plot(df['Date'], df['AQI'], label='AQI', color='blue', linestyle='-', marker='o')
    plt.title('Overall AQI Trend Over Time', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('AQI Value', fontsize=12)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to plot individual pollutant levels (PM2.5, PM10, CO) over time
def plot_pollutant_levels(df):
    plt.figure(figsize=(14, 10))

    # Plot PM2.5 levels
    plt.subplot(3, 1, 1)
    plt.plot(df['Date'], df['PM2.5'], label='PM2.5', color='red', linestyle='-', marker='x')
    plt.title('PM2.5 Trend Over Time')
    plt.xlabel('Date')
    plt.ylabel('PM2.5 Level')
    plt.xticks(rotation=45)

    # Plot PM10 levels
    plt.subplot(3, 1, 2)
    plt.plot(df['Date'], df['PM10'], label='PM10', color='green', linestyle='-', marker='^')
    plt.title('PM10 Trend Over Time')
    plt.xlabel('Date')
    plt.ylabel('PM10 Level')
    plt.xticks(rotation=45)

    # Plot CO levels
    plt.subplot(3, 1, 3)
    plt.plot(df['Date'], df['CO'], label='CO', color='purple', linestyle='-', marker='o')
    plt.title('CO Trend Over Time')
    plt.xlabel('Date')
    plt.ylabel('CO Level')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

# Function to plot AQI values across different dates using a bar plot
def plot_aqi_comparison(df):
    plt.figure(figsize=(10, 6))
    plt.bar(df['Date'], df['AQI'], color='orange', alpha=0.7)
    plt.title('AQI Comparison Across Different Dates', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('AQI Value', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to plot a box plot to analyze the distribution of AQI values
def plot_aqi_distribution(df):
    plt.figure(figsize=(8, 6))
    plt.boxplot(df['AQI'], vert=False, patch_artist=True, notch=True, boxprops=dict(facecolor="skyblue"))
    plt.title('AQI Distribution', fontsize=14)
    plt.xlabel('AQI Value', fontsize=12)
    plt.tight_layout()
    plt.show()

# Function to plot the relationship between AQI and PM2.5 levels using a scatter plot
def plot_aqi_vs_pm25(df):
    plt.figure(figsize=(10, 6))
    plt.scatter(df['PM2.5'], df['AQI'], color='red', alpha=0.5)
    plt.title('Relationship Between PM2.5 and AQI', fontsize=14)
    plt.xlabel('PM2.5 Level', fontsize=12)
    plt.ylabel('AQI Value', fontsize=12)
    plt.tight_layout()
    plt.show()

# Function to plot the overall AQI trend with a customized style
def plot_customized_aqi_trend(df):
    plt.style.use('seaborn-darkgrid')  # Set a predefined style for a polished look
    plt.figure(figsize=(10, 6))
    plt.plot(df['Date'], df['AQI'], label='AQI', color='green', linestyle='-', marker='o')
    plt.title('Customized AQI Trend Over Time', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('AQI Value', fontsize=12)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Main function to load the dataset and call the plotting functions
def main():
    # Load the dataset
    df = pd.read_csv("city_day.csv")
    
    # Convert 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Select relevant columns
    columns_to_plot = ['Date', 'PM2.5', 'PM10', 'CO', 'AQI']
    df_relevant = df[columns_to_plot]

    # Plot each of the different visualizations
    plot_aqi_trend(df_relevant)
    plot_pollutant_levels(df_relevant)
    plot_aqi_comparison(df_relevant)
    plot_aqi_distribution(df_relevant)
    plot_aqi_vs_pm25(df_relevant)
    plot_customized_aqi_trend(df_relevant)

# Run the main function
if __name__ == "__main__":
    main()
```
